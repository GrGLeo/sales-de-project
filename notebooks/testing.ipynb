```python
import requests
import os
import pandas as pd
import datetime
os.chdir('..')
print(os.getcwd())
from utils import flat_json

pd.options.mode.copy_on_write = True
```

```python
url = 'http://localhost:8000/retail'
params = {'limit':5}

response = requests.get(url,params=params)
```

```python
content = response.json()
content = flat_json(content)

df = pd.DataFrame(content)

df.head()
```

```python
df_user = df[['user_name','department','sexe','birth_date']]
df_user.loc[:,'birth_date'] = pd.to_datetime(df_user['birth_date'])
df_user.loc[:,'age'] = df_user['birth_date'].apply(lambda x : datetime.datetime.today().year - x.year)

mapper = {
    'LB_NAME':'user_name',
    'CD_DEPARTMENT':'department',
    'CD_SEXE':'sexe',
    'DT_BIRTH':'birth_date',
    'NB_AGE':'age'
}

df_user.rename({v:k for k,v in mapper.items()}, axis=1)
df_user.drop_duplicates(['LB_NAME','DT_BIRTH'],inplace=True)
```
